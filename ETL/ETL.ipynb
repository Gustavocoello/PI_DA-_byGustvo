# ETL (EXTRACT TRANSFORM LOAD)

```python
# Importamos la librerias 
import pandas as pd 
import numpy as np
```

```python
# traemos los datos 
data = ("Data/homicidios.xlsx")
datos= ("Data/lesiones.xlsx")
```

```python
# cargamos los datos 
data1 = pd.read_excel(data, sheet_name="HECHOS")
data2 =  pd.read_excel(data, sheet_name="VICTIMAS")
data3= pd.read_excel(datos)
```

```python
# COnvertimos la data en Dataframe 
df_facts = pd.DataFrame(data1)
df_vict  = pd.DataFrame(data2)
df_lesion= pd.DataFrame(data3)
```

#### Lesiones Dataset

```python
df_lesion.info()
```

```python
df_lesion.head()
```

```python
# Borraremos las columnas que tenga al menos 5 SD, igual me parece valido dejar las columnas con SD,
# porque asi nos daremos cuenta a que horario se frecuenta estos acontecimientos.
df_lesion_SD = df_lesion.replace("SD", np.nan, inplace=True)

# Definimos el umbral minimo de valores no nulos en la fila
umbral = len(df_lesion.columns) - 5

# Elimina filas que no cumplen con el umbral 
df_filtrado_le = df_lesion.dropna(thresh=umbral)

# Veamos las filas que quedaron 
print(f" Tenemos columnas y filas {df_filtrado_le.shape}")
```

```python
def suma_duplicados(df):
    df_duplicados = pd.DataFrame({
        "Numero de Duplicados" :  df[df.duplicated()].sum(),
        "Porcentaje de Duplicados" : (df[df.duplicated()].sum() / df.shape[0]) * 100.0
    })
    df_duplicados["Porcentaje de Duplicados"] = df_duplicados["Porcentaje de Duplicados"].astype(str) + "%"
    return df_duplicados
```

```python
suma_duplicados(df_filtrado_le)
```

```python
# Veamos si hay valores faltantes
def nulls(df, decimales=2):
    df_nulos = pd.DataFrame({
        "Numero de nulos" : df.isnull().sum(),
        "Porcentaje de nulos" : (df.isnull().sum() / df.shape[0]) * 100.0
    })    
    df_nulos["Porcentaje de nulos"] = df_nulos["Porcentaje de nulos"].round(decimales).astype(str) + "%" 
    return df_nulos 
```

```python
nulls(df_filtrado_le, decimales=2)
```

```python
# Borraremos las columnas con  90% de valores faltantes 
df_filtrado_le = df_filtrado_le.drop(columns=['gravedad', 'otra_direccion'])
```

```python
# Columnas no necesarias 
# Borraremos las columas carro, moto, camion, ciclista, transporte publico
df_filtrado_le = df_filtrado_le.drop(columns=['moto',	'auto', 'transporte_publico',	'camion',	'ciclista'])
```

```python
# Eliminaremos los valores faltantes
df_filtrado_le = df_filtrado_le.dropna(subset=['direccion_normalizada', 'calle', 'altura', 'victima'])
df_filtrado_le.shape
```

```python
df_filtrado_le.columns
```

```python
# Cambiamos el tipo de dato a la columna hora
df_filtrado_le['hora'] = df_filtrado_le['hora'].astype(str)
```

```python
# Para el analisis necesitamos del anio 2016 al 2021, revisemos cuantos anio tenemos 
filtro = df_filtrado_le['aaaa'].unique()
print(filtro)
```

```python
# Reseteamos el inidce 
df_filtrado_le.reset_index()
```

##### Parquet

```python
df_filtrado_le.to_parquet("lesiones.parquet")
```

### Concatenar los archivos lesiones y hechos 
contienen misma columnas mas informacion mejor analisis  

```python
# Creamos una columna nueva llamada LUGAR_DEL_HECHO
df_filtrado_le['LUGAR_DEL_HECHO'] = df_filtrado_le['direccion_normalizada']
```

```python
df_filtrado_le.info()
```

```python
# Cambiamos los nombres de los dataframe df_filtrado_le para que se parezca a df_facts
df_filtrado_le.rename(columns= {
    'id': 'ID',
    'n_victimas'            : 'N_VICTIMAS',
    'fecha'                 : 'FECHA',
    'aaaa'                  : 'AAAA',
    'mm'                    : 'MM',
    'dd'                    : 'DD',
    'hora'                  : 'HORA',
    'franja_hora'           : 'HH',
    'LUGAR_DEL_HECHO'       : 'LUGAR_DEL_HECHO',
    'tipo_calle'            : 'TIPO_DE_CALLE',
    'calle'                 : 'Calle',
    'altura'                : 'Altura',
    'cruce'                 : 'Cruce',
    'direccion_normalizada' : 'Dirección Normalizada',
    'comuna'                : 'COMUNA',
    'geocodificacion_CABA'  : 'XY (CABA)',
    'longitud'              : 'pos x',
    'latutid'               : 'pos y',
    'participantes'         : 'PARTICIPANTES',
    'victima'               : 'VICTIMA',
    'acusado'               : 'ACUSADO'
}, inplace= True )
```
